# Load data

```python
import pandas as pd
ZKI_Innovation = pd.read_csv("ZKI_Innovation_dirty_data.csv")
ZKI_Innovation
```

```python
# new data frame with split value columns 
new = ZKI_Innovation["type_of_flat"].str.split(",", n = 1, expand = True) 
  
# df display 
# making separate first name column from new data frame 
ZKI_Innovation["flat_type"]= new[0] 
  
# making separate last name column from new data frame 
ZKI_Innovation["size"]= new[1] 
new2 = ZKI_Innovation["size"].str.split(",", n = 1, expand = True) 

ZKI_Innovation["size"]= new2[0] 
ZKI_Innovation["level"]= new2[1]  

ZKI_Innovation = ZKI_Innovation.drop(['type_of_flat'], axis=1)
ZKI_Innovation
```

```python
# column cleaning
ZKI_Innovation['level'] = ZKI_Innovation['level'].str.replace("эт.", "", case = False) 
ZKI_Innovation['size'] = ZKI_Innovation['size'].str.replace("м²", "", case = False) 
ZKI_Innovation
```

```python
ZKI_Innovation['level'] = ZKI_Innovation['level'].str.replace("эт.", "", case = False) 
ZKI_Innovation['size'] = ZKI_Innovation['size'].str.replace("м²", "", case = False) 
ZKI_Innovation
```

```python
new3 = ZKI_Innovation["level"].str.split("/", n = 1, expand = True) 

ZKI_Innovation["flat_floor"]= new3[0] 
ZKI_Innovation["builing_floor"]= new3[1] 
ZKI_Innovation = ZKI_Innovation.drop(['level'], axis=1)
ZKI_Innovation
```

```python
new4 = ZKI_Innovation["location"].str.split(",", n = 1, expand = True) 

ZKI_Innovation["street"]= new4[0] 
ZKI_Innovation["number_of_building"]= new4[1] 
ZKI_Innovation = ZKI_Innovation.drop(['location'], axis=1)
ZKI_Innovation
```

```python
new5 = ZKI_Innovation["number_of_building"].str.split(",", n = 1, expand = True) 

ZKI_Innovation["number_of_building"]= new5[0] 
ZKI_Innovation["building_area"]= new5[1] 
ZKI_Innovation
```

```python
ZKI_Innovation = ZKI_Innovation.dropna()
ZKI_Innovation
```

```python
ZKI_Innovation.info()
```

```python
# let's make size numeric
ZKI_Innovation['flat_size'] = pd.to_numeric(ZKI_Innovation['size'], errors='coerce')
ZKI_Innovation = ZKI_Innovation.drop(['size'], axis=1)
ZKI_Innovation
```

#### Regular expressions

```python
import re
re.compile('<title>(.*)</title>')

rooms = []
for h in ZKI_Innovation.flat_type:
    n = [(s) for s in re.findall(r'\d{1}', h)]
    rooms.append(n)
    
roomsT = pd.DataFrame(rooms)
```

```python
# how much rooms ?
roomsT[0].unique()
```

```python
ZKI_Innovation['Rooms'] = roomsT
ZKI_Innovation
```

```python
# number of properties rows
len(ZKI_Innovation['Rooms'])
```

```python
# convert rooms to numeric
ZKI_Innovation['room'] = pd.to_numeric(ZKI_Innovation['Rooms'], errors='coerce')
ZKI_Innovation = ZKI_Innovation.drop(['Rooms'], axis=1)
ZKI_Innovation
```

```python
# how many flats with room in the kitchen?
ZKI_Innovation.room = ZKI_Innovation.room.fillna(0)
ZKI_Innovation
```

```python
r=[]
for x in ZKI_Innovation['room']:
    r.append(int(x))
   #r.append(int(x[1:]))
    
ZKI_Innovation['room'] = r
ZKI_Innovation
```

```python
ZKI_Innovation.info()
```

```python
# Rent proce function
def f(ZKI_Innovation):
    if ZKI_Innovation['room'] == 0:
        val = 30000
    elif ZKI_Innovation['room'] == 1:
        val = 35000
    else:
        val = 40000
    return val
```

```python
ZKI_Innovation['Rent_price'] = ZKI_Innovation.apply(f, axis=1)
ZKI_Innovation
```

#### Calculate ROI

```python
# if we are in Russia we can reduce taxes by renting flat for 11 month, not 12
ZKI_Innovation['ROI'] = (11 * ZKI_Innovation.Rent_price)/ZKI_Innovation.price
ZKI_Innovation
```

```python
ZKI_Innovation['Price/m²'] = ZKI_Innovation['price']/ZKI_Innovation['flat_size']
ZKI_Innovation
```
