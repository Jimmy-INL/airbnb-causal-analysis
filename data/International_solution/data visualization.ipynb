### Hypothesis 1: If I will buy a flat with 1 room in this building, will I pay more money than for a flat with room in the kitchen?

```python
plt.figure(figsize=(15,10))
ax = sns.boxplot(x='room', y = 'price', data = ZKI_Innovation)
ax.tick_params(labelsize=13)
plt.show()
```

#### Hypothesis 2: If I will buy flat with 1 room in this building, will I get maxiumum ROI ?

```python
plt.figure(figsize=(15,10))
ax = sns.boxplot(x='room', y = 'ROI', data = ZKI_Innovation)
ax.tick_params(labelsize=13)
plt.show()
```

```python
areas = ZKI_Innovation.loc[ZKI_Innovation['building_area'].isin(['Москва','Новоивановское'])]
areas 
```

#### Hypothesis 3: In region Х is it cheaper to buy an apartment?

```python
plt.figure(figsize=(15,10))
ax = sns.boxplot(x='building_area', y = 'Price/m²', data = areas)
ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right")
ax.tick_params(labelsize=13)
plt.show()
```

#### Hypothesis 4: Shoould I buy flat in area X ?

```python
plt.figure(figsize=(15,10))
ax = sns.boxplot(x='building_area', y = 'ROI', data = areas)
ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right")
ax.tick_params(labelsize=13)
plt.show()
```
